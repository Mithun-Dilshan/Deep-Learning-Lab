```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(scale=1.0, size=100000)

plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.uniform(size=100000), density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(np.random.normal(size=100000), density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python

import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
import seaborn as sns

pearson_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')

spearman_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation')

plt.show()
```
